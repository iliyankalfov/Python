```python
root = tk.Tk() 
root.title("Calculator") 
root.geometry("300x300")

calculator_label = tk.Label(root, text="Calculator \n", font = ("Helvetica",20)) 
calculator_label.grid(row=0,column=1)

label_entry = tk.Label(root, text="Input:", font=("Helvetica",15))
label_entry.grid(row=2,column=0)

entry = tk.Entry(root, width=20, font=("Arial",20)) 
entry.grid(row=2,column=1)

empty_row_label = tk.Label(root, text="")
empty_row_label.grid(row=3, column=1)

results_list = ["0"]
result_label = tk.Label(root, bd=3, text="")
final_result_label = tk.Label(root, bd=3, text="")

def show_entry_addition(): 
    text = entry.get()
    if not text == "":
        results_list.append(text)
        result_label.configure(text = " = " + 
                               results_list[len(results_list)-2] + " + " + text + " = ",
                               font=("Helvetica",15)) 

        results_list.append(str(int(results_list[len(results_list)-2]) + int(text)))

        result_label.grid(row=4,column=1) 
        entry.delete(0,'end')
    
def show_entry_substraction(): 
    text = entry.get()
    if not text == "":
        results_list.append(text)
        result_label.configure(text = " = " + 
                               results_list[len(results_list)-2] + " - " + text + " = ",
                               font=("Helvetica",15)) 

        results_list.append(str(int(results_list[len(results_list)-2]) - int(text)))

        result_label.grid(row=4,column=1) 
        entry.delete(0,'end')
    
def show_result():
    last_result = results_list[len(results_list)-1]
    final_result_label.configure(text = "Final result: " + last_result, font=("Helvetica",15))
    final_result_label.grid(row=5,column=1)
    entry.configure(state="disabled")
    
    add_button.configure(state="disabled")
    sub_button.configure(state="disabled")
    result_button.configure(state="disabled")

add_button = tk.Button(root, bd=3, text="+", width=3, font=("Helvetica",20), command=show_entry_addition) 
add_button.grid(row=6,column=0)

sub_button = tk.Button(root, bd=3, text="-", width=3, font=("Helvetica",20), command=show_entry_substraction) 
sub_button.grid(row=6, column=1)

result_button = tk.Button(root, bd=3, text="=", width=3, font=("Helvetica",20), command=show_result)
result_button.grid(row=6, column=2)

root.mainloop()
```
